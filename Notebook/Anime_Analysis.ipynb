####  Import Required Packages

```python
#Import required packages

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

#### Explore Dataset

```python
#Load the dataset
df = pd.read_csv(r"C:\Users\LENOVO\Downloads\archive\anime.csv")
#Copy the dataframe
df_new= df.copy()
df_new.head()
```

```python
#Show the columns and their datatype 
df_new .info()
```

####  Line Chart 

```python
# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the data as a line chart
ax.plot(df_new['rating'])

# Add labels and title
ax.set_xlabel('Anime Index')
ax.set_ylabel('Rating')
ax.set_title('Anime Rating Over Index')

# Show the chart
plt.show()
```

 Main goal for this visulizatiion is identify any overall patterns or trends in the data.
 We can see from the line chart whether ratings are generally increasing, decreasing, or staying the same over time.
 This visualization could be improved by adding more context to the plot. For example, we could add a horizontal line to represent the average rating of all anime in the dataset. This would give us a better sense of how individual anime compare to the overall average

###  Bar Chart

```python
# Group data by genre and calculate the mean rating for each genre
genre_data = df_new.groupby('genre').mean()['rating'].sort_values(ascending=False).head(20)

# Create a figure and axis object
fig, ax = plt.subplots()

# Plot the data as a horizontal bar chart
ax.barh(genre_data.index, genre_data.values)

# Add labels and title
ax.set_xlabel('Rating')
ax.set_ylabel('Genre')
ax.set_title('Average Anime Rating by Genre')

# Invert y-axis to show highest ratings at the top
ax.invert_yaxis()

# Show the chart
plt.show()

    
```

The intended goal of this visualization is to show the average rating of anime by genre. This could help us identify which genres tend to have higher ratings and which tend to have lower ratings.

 We can see from the bar chart which genres have the highest average rating and which have the lowest.

The current plot sorts the genres by their average rating in descending order. However, we could also sort the genres alphabetically or by the number of anime in each genre. This would provide a different perspective on the data and allow us to identify any patterns or trends by genre.

### Scatter Plot

```python
# Create a scatter plot with rating on the x-axis and number of members on the y-axis
plt.scatter(df_new['rating'], df_new['members'], alpha=0.5)

# Add labels and title
plt.xlabel('Rating')
plt.ylabel('Number of Members')
plt.title('Anime Ratings vs. Number of Members')

# Show the plot
plt.show()
```

This scatter plot can help us visualize the relationship between anime ratings and the number of members who have watched them.
We can obtain information from plot like range of the number of members who have watched the anime and correlation between anime ratings and the number of member.
It can be improved by Adding a color scheme to differentiate between different types of anime (e.g., TV shows, movies, OVAs)

### Heatmap

```python
# Create a pivot table from the data
pivot_table = df.pivot_table(index='type', columns='genre', values='rating')

# Create a heatmap from the pivot table using Seaborn
sns.heatmap(pivot_table, cmap='coolwarm')

# Set the plot title and axis labels
plt.title('Anime Ratings by Type and Genre')
plt.xlabel('Genre')
plt.ylabel('Type')

# Show the plot
plt.show()
```

This visulation provide a visual representation of the relationship between the anime ratings, types, and genres.
This helps us to know genres and types tend to have higher or lower ratings and to check  any patterns or trends in the data that may be of interest.
this can be improved by adding  annotations to the heatmap to display the actual rating values for each cell, allowing for more precise analysis.
